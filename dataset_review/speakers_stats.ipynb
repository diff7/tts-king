```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../speaker_stats.csv', delimiter='|')
```

```python
data.shape
```

```python
counts = data.groupby('speaker_id').sum('num_sentences').sort_values('num_sentences', ascending=False)['num_sentences']
```

```python
plt.figure(figsize=(3,3))
plt.hist(counts, bins = 20)
plt.title('Number of sentences per speaker')
```

```python
data[data['num_sentences'] > 100].shape
```

```python
data.groupby('speaker_id').sum('num_sentences').sort_values('num_sentences', ascending=True)[:20]['num_sentences'].plot(kind='barh', figsize=(7,7))
plt.title('Least 20 speakers by number of sentences')
```

```python
data.groupby('speaker_id').sum('num_sentences').sort_values('num_sentences', ascending=False)[:20]['num_sentences'].plot(kind='barh', figsize=(7,7))
plt.title('TOP 20 speakers by number of sentences')
```
