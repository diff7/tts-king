```python
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd

from tts_king import TTSKing

from scipy.io.wavfile import write
from logmmse import logmmse
```

```python
tts_king = TTSKing(config_path='config.yaml')
```

```python
def gen_and_save(name,text, path, speaker, duration=1, pitch=1, energy=1, save=False):
    mel_postnet = tts_king.generate_mel(text, 
                                        duration_control=duration,
                                        pitch_control=pitch,
                                        energy_control=energy,
                                        speaker=speaker,)
    wav = tts_king.mel_to_wav(mel_postnet)
    if save:
        save_path = os.path.join(path, '_'.join([speaker, name])+'.wav')
        wav = logmmse(np.array(wav[0][0]), 22050, output_file=save_path, initial_noise=1, window_size=150, noise_threshold=0.15)
        write(save_path, 22050, wav.astype(np.int16))
    return wav[0][0]
```

```python
path = './'

speakers = ['Rezalin_Aleksandr_abooks_voxforge','Goblin_abooks_voxforge','hajdurova_ailab','user4_mozilla','nikolaev_ailab']

text = 'босс я же щас расплачусь'

wav = gen_and_save(name='cry',text=text, path=path, speaker=speakers[1], duration=1, pitch=1.3, energy=1)
ipd.Audio(wav, rate=22050)
```
